```python
# 평균 계산 함수: 사용자로부터 임의 개수의 숫자를 입력받고, 이들의 평균을 계산하여 반환하는 함수를 작성하세요.

def get_numbers():
    my_list = []
    while True:
        input_number = input('입력: ')

        if input_number == 'q':
            break
        else:
            my_list.append(int(input_number))
    return my_list

def calculate_mean(my_list):
    total = sum(my_list)
    mean = total / len(my_list)
    return mean

my_list = get_numbers()
mean = calculate_mean(my_list)
print(mean)

```

```python
# 최대값 찾기 함수: 사용자로부터 임의 개수의 숫자를 입력받고, 그 중 가장 큰 값을 찾아서 반환하는 함수를 작성하세요.
def get_numbers():
    my_list = []
    while True:
        input_number = input('입력: ')

        if input_number == 'q':
            break
        else:
            my_list.append(int(input_number))
    return my_list

def return_max_val(my_list):
    return max(my_list)

def return_min_val(my_list):
    return min(my_list)

my_list = get_numbers()
max_val = return_max_val(my_list)
print(max_val)
```

```python
# 짝수 판별 함수: 짝수의 경우 True, 홀수의 경우 False를 반환하는 함수를 작성하세요.

def check(n):
    if n % 2 == 0:
        return True
    else:
        return False

n = 8
result = check(n)
print(result)

```
