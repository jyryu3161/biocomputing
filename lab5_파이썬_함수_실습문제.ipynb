## 외장함수 예시

```python
import copy

a = [1,2,3,4,5]
b = a
b[2] = 100
print(a)
print(b)

a = [1,2,3,4,5]
b = copy.deepcopy(a)
b[2] = 100
print(a)
print(b)
```
