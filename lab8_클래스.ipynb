## 클래스란?

*   클래스는 객체를 정의하는 틀
*   객체는 클래스의 인스턴스(instance)

```python

class Dog:
    # 클래스 속성
    species = 'Canis familiaris'

    # 초기화 메서드
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 메서드
    def bark(self):
        return '%s says woof!'%(self.name)

# 객체 생성
my_dog = Dog('Rex', 5)

# 객체의 속성과 메서드 사용
print(my_dog.name)  # 출력: Rex
print(my_dog.age)   # 출력: 5
print(my_dog.bark()) # 출력: Rex says woof!

print(Dog.species)
```

요약
* class 키워드로 클래스를 정의합니다.
*   __init__ 메서드는 초기화 메서드로, 객체가 생성될 때 호출됩니다.
*   self는 인스턴스 자신을 참조합니다.




## 클래스 속성과 인스턴스 속성

### 클래스 속성
* 클래스 자체에 속하는 속성
* 모든 인스턴스가 공유
### 인스턴스 속성
* 각 인스턴스에 속하는 속성
* 각 인스턴스마다 독립적

```python
class Dog:
    species = 'Canis familiaris'  # 클래스 속성

    def __init__(self, name, age):
        self.name = name  # 인스턴스 속성
        self.age = age  # 인스턴스 속성

dog1 = Dog('Buddy', 3)
dog2 = Dog('Molly', 5)

print(dog1.species)  # 출력: Canis familiaris
print(dog2.species)  # 출력: Canis familiaris

print(dog1.name)  # 출력: Buddy
print(dog2.name)  # 출력: Molly
```

## 메서드

### 인스턴스 메서드
* 클래스 내부에 정의된 함수
* 인스턴스에서 호출

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return '%s says woof!'%(self.name)

my_dog = Dog('Rex', 5)
print(my_dog.bark())  # 출력: Rex says woof!
```

### 클래스 메서드
* 클래스에 속하는 메서드
* @classmethod 데코레이터를 사용

```python
class Dog:
    species = 'Canis familiaris'

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def get_species(cls):
        return cls.species

print(Dog.get_species())  # 출력: Canis familiaris
```

### 정적 메서드
* 클래스에 속하지만, 인스턴스나 클래스 속성에 접근하지 않음
* @staticmethod 데코레이터를 사용

```python
class Dog:
    @staticmethod
    def bark():
        return 'Woof!'

print(Dog.bark())  # 출력: Woof!
```

## 상속
* 새로운 클래스가 기존 클래스를 물려받는 것
* 코드 재사용성을 높임

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return '%s says woof!'%(self.name)

class Cat(Animal):
    def speak(self):
        return'%s says meow!'%(self.name)

dog = Dog('Rex')
cat = Cat('Whiskers')

print(dog.speak())  # 출력: Rex says woof!
print(cat.speak())  # 출력: Whiskers says meow!
```

### 메서드 오버라이딩
* 부모 클래스의 메서드를 자식 클래스에서 재정의하는 것

```python
class Animal:
    def speak(self):
        return 'Some sound'

class Dog(Animal):
    def speak(self):
        return 'Woof!'

dog = Dog()
print(dog.speak())  # 출력: Woof!
```

### self 필요성 - self는 클래스 인스턴스 자신을 가리키는 변수

* "self"는 클래스의 인스턴스 변수와 메서드에 접근하기 위해 필요
* 인스턴스 변수는 객체마다 다를 수 있으므로, "self"를 사용하여 해당 객체의 변수를 참조

```python
class Person:
    def __init__(self, name, age):
        self.name = name  # 인스턴스 변수 name
        self.age = age    # 인스턴스 변수 age

    def introduce(self):
        return 'Hi, my name is %s and I am %s years old.'%(self.name, self.age)

# 객체 생성
person1 = Person('Alice', 30)
person2 = Person('Bob', 25)

# 메서드 호출
print(person1.introduce())  # 출력: Hi, my name is Alice and I am 30 years old.
print(person2.introduce())  # 출력: Hi, my name is Bob and I am 25 years old.
```

BankAccount 클래스 정의

```python
# ==========================
# BankAccount 클래스 정의
# ==========================
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        """입금 메서드"""
        if amount <= 0:
            print("입금 금액은 0보다 커야 합니다.")
        else:
            self.balance += amount
            print(f"{amount}원이 입금되었습니다.")

    def withdraw(self, amount):
        """출금 메서드"""
        if amount > self.balance:
            print("잔액이 부족합니다.")
        else:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다.")

    def show_balance(self):
        """잔액 확인 메서드"""
        print(f"{self.owner}님의 현재 잔액은 {self.balance}원입니다.")


# ==========================
# 객체 생성 및 메서드 호출
# ==========================
# 1️⃣ 계좌 객체 생성
account1 = BankAccount("홍길동", 10000)
account2 = BankAccount("김영희", 5000)

# 2️⃣ 입금 / 출금 / 잔액조회
account1.deposit(2000)
account1.withdraw(8000)
account1.show_balance()

# 3️⃣ 다른 계좌에서도 수행
account2.deposit(10000)
account2.withdraw(3000)
account2.show_balance()

# 4️⃣ 여러 계좌 관리 리스트
accounts = [account1, account2]
for acc in accounts:
    acc.show_balance()
```

학생 클래스 정의

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.scores = []

    def add_score(self, score):
        """점수 추가"""
        self.scores.append(score)
        print(f"{self.name} 학생의 점수 {score}점이 추가되었습니다.")

    def get_average(self):
        """평균 계산"""
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)

    def __str__(self):
        """객체 정보를 문자열로 표현"""
        return f"이름: {self.name}, 학번: {self.student_id}, 평균: {self.get_average():.2f}"


# ==========================
# 객체 생성 및 메서드 호출
# ==========================
# 1️⃣ 학생 객체 생성
student1 = Student("이민수", "2025001")
student2 = Student("박지현", "2025002")

# 2️⃣ 점수 추가
student1.add_score(90)
student1.add_score(85)
student2.add_score(70)
student2.add_score(95)
student2.add_score(88)

# 3️⃣ 평균 계산 및 출력
print(student1)
print(student2)

# 4️⃣ 평균 비교
students = [student1, student2]
top_student = max(students, key=lambda s: s.get_average())
print(f"가장 높은 평균을 가진 학생은 {top_student.name} ({top_student.get_average():.2f}점)")
```

## 문제

- 도서관에서 책 정보를 관리하려고 합니다. 책의 제목(title), 저자(author), 출판연도(year)를 속성으로 가지는 Book 클래스를 만들어보세요.


구현 조건

- 클래스명: Book
- 생성자 (__init__)
- 매개변수: title, author, year
- 속성으로 저장: self.title, self.author, self.year

메서드 3개 구현
- show_info() | 책의 정보를 "제목: ○○, 저자: ○○, 출판연도: ○○" 형태로 출력 |
- update_year(new_year) | 출판연도를 새로운 값으로 변경 하고 “출판연도가 변경되었습니다.” 출력 |
- is_recent() | 출판연도가 2019년 이후이면 True 반환, 그 외에는 False 반환 |

실행 예시


```python
# Book 클래스 작성 후 아래 코드로 테스트해보세요.

class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def show_info(self):
        print("제목: %s, 저자: %s, 출판연도: %s"%(self.title, self.author, self.year))

    def update_year(self, new_year):
        self.year = new_year
        print("출판연도가 변경되었습니다.")

    def is_recent(self):
        if self.year >= 2019:
            return True
        else:
            return False


book1 = Book("파이썬 입문", "홍길동", 2018)
book2 = Book("인공지능 개론", "이순신", 2023)

book1.show_info()
book2.show_info()

book1.update_year(2021)
book1.show_info()

print(book1.is_recent())   # True
print(book2.is_recent())   # True
```
